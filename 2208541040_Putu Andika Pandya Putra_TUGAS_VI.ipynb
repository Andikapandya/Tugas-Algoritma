# Mengimport Library

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Memanggil data yang akan dibaca

```python
df_I = pd.read_csv ("bunga_iris.csv")
print (df_I)
```

```python
df_II = pd.read_csv ("bunga_iris_target.csv")
print (df_II)
```

```python
df_III = pd.read_csv ("nama_bunga.csv")
print (df_III)
```

# Mengecek Kesesuaian Data

```python
df_I.info()
```

```python
df_II.info()
```

```python
df_III.info()
```

# Menggabungkan Data

```python
#menggabungkan data I,II, dan III dengan cara menambahkan column 
df_I["flower_name"]=df_III["flower_name"]
```

```python
df_I.head()
```

```python
df_I["target"]=df_II["target"]
```

```python
df_I.head()
```

```python
#copy wadah
df = df_I.copy()
```

```python
print (df)
```

# Mengurutkan Data bunga_iris

```python
#mengurutkan data dari terkecil ke terbesar
df.sort_values("sepal length (cm)")
```

```python
#mengurutkan data dari terbesar ke terkecil
df.sort_values("sepal length (cm)",ascending=0)
```

# Menganalisis Data

```python
#mencari modus pada setiap column
for col_name in ["sepal length (cm)","sepal width (cm)","petal length (cm)","petal width (cm)","flower_name","target"]:
  print(df[col_name].value_counts(), '\n')
```

```python
#menghitung besaran dari sepal
df["ukuran sepal (cm^2)"] = df["sepal length (cm)"]*df["sepal width (cm)"]
```

```python
df.head()
```

```python
#menghitung besaran petal
df["ukuran petal (cm^2)"] = df["petal length (cm)"]*df["petal width (cm)"]
```

```python
df.head()
```

```python
#mencari nilai min dari masing masing nama bunga
df.groupby(["flower_name"]).min()
```

```python
#mencari nilai max dari masing masing nama bunga
df.groupby(["flower_name"]).max()
```

```python
#membandingkan diantara nama bunga yang mana lebih besar
df.groupby(["flower_name"]).sum()
```

```python
#mencari rata rata besaran masing masing nama bunga
df.groupby(["flower_name"]).mean()
```

```python
df.info()
```

# Visualisasi Data

```python
#melihat grafik nama bunga berdasarkan ukuran sepal
plt.rcParams["figure.figsize"]=(15,10)
df.plot(x = 'flower_name',y = 'ukuran sepal (cm^2)')
plt.title('SEPAL BUNGA')
plt.xlabel('flower_name')
plt.ylabel('ukuran sepal (cm^2)')
```

```python
#melihat grafik nama bunga berdasarkan ukuran petal
plt.rcParams["figure.figsize"]=(10,5)
df.plot(x = 'flower_name',y = 'ukuran petal (cm^2)')
plt.title('PETAL BUNGA')
plt.xlabel('flower_name')
plt.ylabel('ukuran petal (cm^2)')
```

```python
#membuat nilai minimum antara sepal dan petal bunga
jumlah = df.groupby('flower_name', as_index=False)['ukuran sepal (cm^2)', 'ukuran petal (cm^2)'].min()
jumlah.head()
```

```python
plt.rcParams["figure.figsize"]=(10,5)
plt.plot('flower_name', 'ukuran sepal (cm^2)', 'ukuran petal (cm^2)', data=jumlah)
plt.title('nilai minimum antara sepal dan petal bunga')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
#biru = sepal
#kuning = petal
```

```python
#membuat nilai maksimum antara sepal dan petal bunga
jumlah = df.groupby('flower_name', as_index=False)['ukuran sepal (cm^2)', 'ukuran petal (cm^2)'].max()
jumlah.head()
```

```python
plt.rcParams["figure.figsize"]=(10,5)
plt.plot('flower_name', 'ukuran sepal (cm^2)', 'ukuran petal (cm^2)', data=jumlah)
plt.title('nilai maksimum antara sepal dan petal bunga')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
#biru = sepal
#kuning = petal
```

```python
#membuat perbandingan antara ukuran sepal dan petal
jumlah = df.groupby('flower_name', as_index=False)['ukuran sepal (cm^2)', 'ukuran petal (cm^2)'].sum()
jumlah.head()
```

```python
plt.rcParams["figure.figsize"]=(10,5)
plt.plot('flower_name', 'ukuran sepal (cm^2)', 'ukuran petal (cm^2)', data=jumlah)
plt.title('perbandingan sepal dan petal')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
#biru = sepal
#kuning = petal
```

```python
jumlah = df.groupby('flower_name', as_index=False)['ukuran sepal (cm^2)', 'ukuran petal (cm^2)'].mean()
jumlah.head()
```

```python
plt.rcParams["figure.figsize"]=(10,5)
plt.plot('flower_name', 'ukuran sepal (cm^2)', 'ukuran petal (cm^2)', data=jumlah)
plt.title('perbandingan mean antara sepal dan petal bunga')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
#biru = sepal
#kuning = petal
```
